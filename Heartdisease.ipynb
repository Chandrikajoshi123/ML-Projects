```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Data collection and preprocessing**

```python
# Loading the csv data to a pandas dataframe
df=pd.read_csv('/content/heart_disease_data.csv')
```

```python
# Printing first five rows of the data set
df.head()
```

```python
#Printing last five rows of the dataset
df.tail()
```

```python
# Number of rows and columns in the dataset
df.shape
```

```python
#Getting some info about the data
df.info()
```

```python
# checking for missing values
df.isnull().sum()
```

```python
#statistical measures about the data
df.describe()
```

```python
# checking the distribution of ttarget variable
df['target'].value_counts()
```

 1--> Defective Heart

 0--> Healthy Heart

**Splitting the features and target**

```python
x=df.drop(columns='target',axis=1)
y=df['target']
print(x)
```

```python
print(y)
```

**Spliting the data into Training data & Test data**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

**Model Training**

**Logistic regrassion model**

```python
model1=LogisticRegression()
```

```python
#Training the LogisticRegrssion model with training data
model1.fit(x_train,y_train)
```

Model Evaluation

Accuracy Score

```python
#Accuracy on training data
x_train_prediction=model1.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy on Training data :", training_data_accuracy)
```

```python
#Accuracy on test data
x_test_prediction=model1.predict(x_test)
testining_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy on Test data :", testining_data_accuracy)
```

**Building a Predictive system**

```python
input_data=(41,0,1,130,204,0,0,172,0,1.4,2,0,2)

#change input data to a numpy array
input_data_as_np_arr=np.asarray(input_data)

#reshape  the numpy array as we are predicting for only on instance
input_data_reshape=input_data_as_np_arr.reshape(1,-1)

prediction=model1.predict(input_data_reshape)
print(prediction)
if prediction[0]==0:
  print("The person does not have a Heart Disease ðŸ˜Š ")
else:
  print(" The Person has Heart Disease ðŸ˜” ")



```
